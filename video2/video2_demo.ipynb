## 1. Variables and types

A variable is a name you give to a value. In Python, there are four types you'll
use constantly: strings, numbers, lists, and dictionaries.


### Strings — text data

Text goes in quotes. Single or double quotes both work, just be consistent.


````python
hole_id = "DDH-001"
project_name = "Greenstone Belt East"

print(hole_id)
print(project_name)
````

### Numbers — Python figures out int vs float

Whole numbers are integers (`int`). Decimal numbers are floats (`float`). Python
sorts this out automatically — you'll almost never have to think about it.


````python
azimuth = 45
dip = -60
grade_gpt = 4.7

print(azimuth, dip, grade_gpt)
````

### Lists — ordered collections

Lists hold multiple values in order. Square brackets, comma-separated.

**Important:** Python lists are zero-indexed. The first item is `[0]`, not `[1]`.
This trips up everyone at first.


````python
grades = [0.05, 0.18, 1.24, 4.78, 6.32, 2.41, 0.31]

print(grades[0])     # first element
print(grades[-1])    # last element
print(len(grades))   # how many items
````

### Dictionaries — labeled collections

Use a dictionary when each piece of data has a label. Curly braces, with
`"label": value` pairs.

Lists are for things in order. Dictionaries are for things by name.


````python
hole = {
    "id": "DDH-001",
    "azimuth": 45,
    "dip": -60,
    "length_m": 50.0,
}

print(hole["id"])
print(hole["azimuth"])
print(hole["length_m"])
````

## 2. Operations and logic

Now we make variables interact: arithmetic, formatted strings, comparisons,
and decisions with `if`.


### Math is what you'd expect


````python
length_m = 4.5
contained_metal_gm = grade_gpt * length_m
half_grade = grade_gpt / 2

print(contained_metal_gm)
print(half_grade)
````

### String formatting with f-strings

Putting an `f` before a string lets you drop variables directly into it using
curly braces. The `:.2f` formats a number to two decimal places.


````python
print(f"Hole {hole_id} hit {grade_gpt} g/t Au")
print(f"Contained gold: {grade_gpt * 4.5:.2f} gram-metres")
````

### Comparisons return True or False

Note the **double** equals sign for comparison. Single equals assigns a value;
double equals checks if two things are equal.


````python
print(grade_gpt > 1.0)
print(grade_gpt == 5.0)
print(hole_id == "DDH-001")
````

### `if` statements — different code depending on what's true

Read it like English: *if this is true, do that. Otherwise if this is true, do
that. Otherwise do this.*


````python
if grade_gpt > 5.0:
    print("High grade")
elif grade_gpt > 1.0:
    print("Mineralized")
else:
    print("Below cutoff")
````

## 3. Loops

Any time you find yourself copy-pasting the same line of code with one thing
changed, you should be writing a loop instead.


### The basic pattern: do something to each item


````python
grades = [0.05, 0.18, 1.24, 4.78, 6.32, 2.41, 0.31]

for grade in grades:
    print(grade)
````

The word `grade` inside the loop is just a variable name I picked. It could be
anything. Python uses whatever name you give it.

### A more useful loop: classify each grade


````python
for grade in grades:
    if grade > 1.0:
        print(f"{grade} g/t — mineralized")
    else:
        print(f"{grade} g/t — below cutoff")
````

### Building up a result: count the mineralized intervals


````python
mineralized_count = 0

for grade in grades:
    if grade > 1.0:
        mineralized_count = mineralized_count + 1

print(f"{mineralized_count} mineralized intervals out of {len(grades)}")
````

### Looping through more complex data


````python
holes = [
    {"id": "DDH-001", "depth_m": 50.0},
    {"id": "DDH-002", "depth_m": 45.0},
    {"id": "DDH-003", "depth_m": 45.0},
]

for hole in holes:
    print(f"{hole['id']} drilled to {hole['depth_m']} m")
````

## 4. Functions

A function is a named bundle of instructions. You define it once. You can use
it as many times as you want.

The `def` keyword starts a function. The name is whatever you want. Things in
parentheses are the inputs. Everything indented underneath is the function
body. The `return` line is what comes back when you use it.


````python
def classify_grade(grade_gpt):
    if grade_gpt > 5.0:
        return "high grade"
    elif grade_gpt > 1.0:
        return "mineralized"
    else:
        return "below cutoff"

# Use it
print(classify_grade(4.78))
print(classify_grade(0.18))
print(classify_grade(6.32))
````

### A more useful function: length-weighted mean grade

A 10 m interval at 1 g/t should count twice as much as a 5 m interval at 1 g/t
when computing an average. This is length-weighting.

The triple-quoted string at the top is a **docstring** — a comment that
explains what the function does. Future-you will be grateful for these.


````python
def weighted_mean_grade(grades, lengths):
    """Return the length-weighted mean of a series of grades.

    A 10 m interval at 1 g/t counts twice as much as a 5 m interval at 1 g/t.
    """
    total_metal = 0
    total_length = 0
    for grade, length in zip(grades, lengths):
        total_metal = total_metal + (grade * length)
        total_length = total_length + length
    return total_metal / total_length

# Try it on some intervals
sample_grades = [0.05, 4.78, 6.32, 0.31]
sample_lengths = [3.0, 4.5, 5.5, 6.5]

avg = weighted_mean_grade(sample_grades, sample_lengths)
print(f"Length-weighted mean: {avg:.2f} g/t")
````

`zip` pairs up two lists element by element: the first grade with the first
length, the second with the second, and so on. It's a common pattern when you
have two lists that go together.


## 5. A few habits worth picking up

These aren't rules. They're habits that compound. Apply them when they feel
natural, ignore them when they don't.

### Name things meaningfully

The biggest one. Future-you reading this in six months will appreciate it.
Especially if future-you is a colleague.


````python
# Hard to read
g = 4.7
x = g * 4.5

# Easy to read
grade_gpt = 4.7
contained_metal_gm = grade_gpt * 4.5
````

### Comment WHY, not WHAT

The code already tells you what it does. Comments are for the reasoning the
code can't show.


````python
# Bad — restates the obvious
contained = grade_gpt * length_m  # multiply grade by length

# Better — explains why
contained = grade_gpt * length_m  # length-weighting for composite grade
````

### If you find yourself copy-pasting code, you probably want a loop or a function

You already saw this in Sections 3 and 4. Looping instead of copy-pasting.
Wrapping logic in a function instead of repeating it.

This principle has a name in software circles — DRY, "don't repeat yourself" —
but the name matters less than the habit.


## 6. Pulling it together on the assay data

Let's bring back the drillhole dataset from Video 1 and apply what we just learned.

This time we'll load the CSV directly from GitHub — no need to upload anything.
`pandas` can read a URL just like a local file.


````python
import pandas as pd

# Load assays.csv directly from the GitHub repo
url = "https://raw.githubusercontent.com/DHeasmanGDS/geodatascience-python-intro/main/video2/assays.csv"
df = pd.read_csv(url)

df.head()
````

That's a small but useful trick. `pd.read_csv` can take a local filename or a URL,
and it figures out what to do. Handy when you're sharing notebooks.

Now apply the `classify_grade` function we wrote earlier to every row:


````python
# Apply our function to every row in the Au_gpt column
df["category"] = df["Au_gpt"].apply(classify_grade)

# Look at the result
df[["HoleID", "From_m", "To_m", "Au_gpt", "category"]]
````

````python
# Count how many intervals fall in each category
df["category"].value_counts()
````

*we wrote ourselves* to classify every interval. Counted them. That's the
whole loop.
